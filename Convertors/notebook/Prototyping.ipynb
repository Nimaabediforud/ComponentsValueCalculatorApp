```python
# All colors used for value and decimal multiplier
colors = {
    "black": 0,
    "brown": 1,
    "red": 2,
    "orange": 3,
    "yellow": 4,
    "green": 5,
    "blue": 6,
    "purple": 7,
    "gray": 8,
    "white": 9,
    "gold": -1,
    "silver": -2
}

# All colors used for tolerance
tolerance = {
    "brown": '1',
    "red": '2',
    "green": '0.5',
    "blue": '0.25',
    "purple": '0.1',
    "gold": '5',
    "silver": '10',
    "": '20' # No color
}

# All colors used for temperature coefficient
temp_coeff = {
    "brown": 100,
    "red": 50,
    "orange": 15,
    "yellow": 25,
}

# All letters used in smd resistor labels
smd_letters = {
    "r": ".",
    "d": 0.5,
    "f": 1.0,
    "g": 2.0,
    "h": 5.0
}

# Global list and variable
spec_col = ["gold", "silver"]
NO_COLOR = ""

```

```python
"""
Validation Functions:
    .
    .
    .
"""

def Validate_No_Color(digit=int, lst=list):
    if NO_COLOR in lst[0:digit]:
        return True


def Validate_Spec_Colors(digit=int, lst=list):
    l = [True for c in spec_col if c in lst[0:digit]]
    if any(l): 
        return True
    

def Validate_Tolerance(digit=int, lst=list):
    if lst[digit] not in tolerance:
        return True
    
    
def Validate_Temp_Coeff(digit=int, lst=list):
    if lst[digit] not in temp_coeff:
        return True
    

```

```python
def Unit_Convertor(num=int):
    """
    
    """
    
    num_of_digits = len(str(num))

    if 0 <= num_of_digits <= 3:
        return f"{num} "
    elif 3 < num_of_digits <= 6:
        num = num/(10**3)
        return f"{num} k"
    elif 6 < num_of_digits <= 9:
        num = num/(10**6)
        return f"{num} M"
    elif 9 < num_of_digits <= 12:
        num = num/(10**9)
        return f"{num} G"
    elif 12 < num_of_digits <= 15:
        num = num/(10**12)
        return f"{num} T"

#test
num = 100000000000000
r = Unit_Convertor(num)
print(r)
```

```python
def Jumper(lbl=[str, list]):
    """
    
    """
    if type(lbl) == list:
        lbl = lbl[0]

    lbl = lbl.lower()
    
    if lbl == "black":
        return "A JUMPER so 0 Ω", "Single band black color"
    elif lbl == "o" or lbl == "0":
        return "A JUMPER so 0 Ω", "Single character label: 'o'"
    elif lbl == "000":
        return "A JUMPER so 0 Ω", "Your label: '000'"
    else:
        return False, False


# test
lst = ['black']
lbl = "O"
l, n = Jumper(lst)
l2, n2 = Jumper(lbl)
print(l, n, sep="\n")
print(l2, n2, sep="\n")
```

```python
def FourColor(lst=list):
    """
    
    """

    #--------------------------------------------------------

    # Initial check for a 4-band resistor code - (Becomes a 3-band if its tolerance is 20%)
    if len(lst) == 3:
        band = lst[0:3] + [""] # Three digits + One no color or tolerance
    elif len(lst) == 4: 
        band = lst[0:4] # Take exactly four bands (digits)
    else:
        return False
    
    #--------------------------------------------------------

    # Check user's input to validate it by validation functions
    check_1 = Validate_No_Color(3, band)
    check_2 = Validate_Spec_Colors(2, band)
    check_3 = Validate_Tolerance(-1, band)

    # Messages in case of invalidation
    if check_1:
        return "## The first three bands must not be COLORLESS!"
    if check_2:
        return "## The first two bands can not be GOLD or SILVER!"
    if check_3:
        return "## The correspondent band to tolerance (last band) is incorrect!"
    
    #--------------------------------------------------------

    # Build up each digit from the 4-band code
    first_two_digits = int("".join([str(colors.get(c)) for c in band[0:2]]))
    third_digit = 10**(colors.get(band[2]))
    fourth_digit = tolerance.get(band[-1])

    #--------------------------------------------------------

    # Calculate and Interpret the final result
    result = f"{Unit_Convertor(first_two_digits*third_digit)}Ω ± {fourth_digit}%"

    #--------------------------------------------------------
    
    return result, band

# test
lst = ['brown', 'black', 'green']
l, n = FourColor(lst)
print(l, n, sep="\n")
```

```python
def FiveColor(lst=list):
    """
    
    """

    #--------------------------------------------------------

    # Initial check for a 5-band resistor code
    if len(lst) == 5: 
        band = lst[0:5] # Take exactly five bands (digits)
    else:
        return False
    
    #--------------------------------------------------------

    # Check user's input to validate it by validation functions
    check_1 = Validate_No_Color(4, band)
    check_2 = Validate_Spec_Colors(3, band)
    check_3 = Validate_Tolerance(-1, band)

    # Messages in case of invalidation
    if check_1:
        return "## The first four bands must not be COLORLESS!"
    if check_2:
        return "## The first three bands can not be GOLD or SILVER!"
    if check_3:
        return "## The correspondent band to tolerance (last band) is incorrect!"
    
    #--------------------------------------------------------

    # Build up each digit from the 5-band code
    first_three_digits = int("".join([str(colors.get(c)) for c in band[0:3]]))
    fourth_digit = 10**(colors.get(band[3]))
    fifth_digit = tolerance.get(band[-1])

    #--------------------------------------------------------

    # Calculate and Interpret the final result
    result = f"{Unit_Convertor(first_three_digits*fourth_digit)}Ω ± {fifth_digit}%"

    #--------------------------------------------------------
    
    return result, band


# test
lst = ['blue', 'purple', 'green', 'brown', 'gold']
l, n = FiveColor(lst)
print(l, n, sep="\n")
```

```python
def SixColor(lst=list):
    """
    
    """

    #--------------------------------------------------------

    # Initial check for a 6-band resistor code
    if len(lst) == 6: 
        band = lst[0:6] # Take exactly six bands (digits)
    else:
        return False
    
    #--------------------------------------------------------

    # Check user's input to validate it by validation functions
    check_1 = Validate_No_Color(4, band)
    check_2 = Validate_Spec_Colors(3, band)
    check_3 = Validate_Tolerance(4, band)
    check_4 = Validate_Temp_Coeff(-1, band)

    # Messages in case of invalidation
    if check_1:
        return "## The first four bands must not be COLORLESS!"
    if check_2:
        return "## The first three bands can not be GOLD or SILVER!"
    if check_3:
        return "## The correspondent band to tolerance (fifth band) is incorrect!"
    if check_4:
        return "## The correspondent band to temperature coefficient is incorrect!"
    
    #--------------------------------------------------------

    # Build up each digit from the 6-band code
    first_three_digits = int("".join([str(colors.get(c)) for c in band[0:3]]))
    fourth_digit = 10**(colors.get(band[3]))
    fifth_digit = tolerance.get(band[4])
    sixth_digit = temp_coeff.get(band[-1])

    #--------------------------------------------------------

    # Calculate and Interpret the final result
    result = f"{Unit_Convertor(first_three_digits*fourth_digit)}Ω ± {fifth_digit}% | TC: {sixth_digit} ppm"

    #--------------------------------------------------------
    
    return result, band


# test
lst = ['red', 'purple', 'blue', 'black', 'brown', 'red']
l, n = SixColor(lst)
print(l, n, sep="\n")

```

```python
def Dip_Res():
    """
    
    """
    #--------------------------------------------------------
    
    col = input(f"Enter the band colors (Separated with '-'): ").strip().lower()
    colors_lst = col.split("-")
    for c in colors_lst:
        if c not in colors.keys():
            return "Oops, Seems like we have a problem!", "Make sure you have not misspelled the name of the colors!"

    #--------------------------------------------------------

    if Jumper(colors_lst) != False:
        return f"Colors: {Jumper(col)[1]}", f"Result: {Jumper(col)[0]}"  
    elif FourColor(colors_lst) != False:
        return f"Colors: {FourColor(colors_lst)[1]}", f"Result: {FourColor(colors_lst)[0]}"  
    elif FiveColor(colors_lst) != False:
        return f"Colors: {FiveColor(colors_lst)[1]}", f"Result: {FiveColor(colors_lst)[0]}"    
    elif SixColor(colors_lst) != False:
        return f"Colors: {SixColor(colors_lst)[1]}", f"Result: {SixColor(colors_lst)[0]}"    
```

```python
def Three_Digit(label=str):
    """
    
    """
    
    #--------------------------------------------------------
    
    if len(label) != 3:
        return False
    
    #--------------------------------------------------------

    r_count = label.count("r")
    
    if  r_count > 1:
        return "Only one letter of R is allowed!", "Please try again!"
    elif r_count == 1:
        if label.index("r") == 0:
            result = f"{'0' + label.replace('r', smd_letters.get('r'))} Ω"
        elif label.index("r") == 1:
            result = f"{label.replace('r', smd_letters.get('r'))} Ω"
        else:
            result = f"{label[0:2]} Ω"
    else:
        # Build up each digit from the 4-band code
        first_two_digits = int(label[0:2])
        third_digit = 10**(int(label[-1]))

        # Calculate and Interpret the final result
        result = f"{Unit_Convertor(first_two_digits*third_digit)}Ω"


    return result, label.upper()
        
    

# test
label = "5r9"
r = Three_Digit(label)
r
```

```python
def Four_Digit(label=str):
    """
    
    """

    #--------------------------------------------------------

    if len(label) != 4:
        return False
    
    #--------------------------------------------------------
    
    if label[3] in list(smd_letters.keys())[1:]:
        r_count = label.count("r")

        if  r_count > 1:
            return "Only one letter of R is allowed!", "Please try again!"
        elif r_count == 1:
            if label.index("r") == 0:
                result = f"{'0' + label[0:3].replace('r', smd_letters.get('r'))} Ω ± {smd_letters.get(label[-1])}%"
            elif label.index("r") == 1 or label.index("r") == 2:
                result = f"{label[0:3].replace('r', smd_letters.get('r'))} Ω ± {smd_letters.get(label[-1])}%"
            else:
                result = f"{label[0:3]} Ω ± {smd_letters.get(label[-1])}%"
        else:
            # Build up each digit from the label
            first_two_digits = int(label[0:2])
            third_digit = 10**(int(label[2]))

            # Calculate and Interpret the final result
            result = f"{Unit_Convertor(first_two_digits*third_digit)}Ω ± {smd_letters.get(label[-1])}%"
        
    elif label[3] == list(smd_letters.keys())[0] or label[3].isdigit():
        r_count = label.count("r")

        if  r_count > 1:
            return "Only one letter of R is allowed!", "Please try again!"
        elif r_count == 1:
            if label.index("r") == 0:
                result = f"{'0' + label.replace('r', smd_letters.get('r'))} Ω"
            elif label.index("r") == 1 or label.index("r") == 2:
                result = f"{label.replace('r', smd_letters.get('r'))} Ω"
            else:
                result = f"{label[0:3]} Ω"
        else:
            # Build up each digit from the label
            first_three_digits = int(label[0:3])
            fourth_digit = 10**(int(label[-1]))

            # Calculate and Interpret the final result
            result = f"{Unit_Convertor(first_three_digits*fourth_digit)}Ω"
        
        
    return result, label.upper()



lbl = "1r0d"
l = Four_Digit(lbl)
l
```

```python
def SMD_Res():
    """
    
    """

    #--------------------------------------------------------

    label = input(f"Enter the label: ").strip().lower()

    check_ = []
    if len(label) == 1 or 3 <= len(label) <= 4:
        for c in label:
            if c.isdigit() or c in ["r", "o"]:
                check_.append(True)
            elif c in smd_letters.keys() and label.index(c) == 3:
                check_.append(True)
            else:
                check_.append(False)
    else:
        return "There should be 3-4 characters in the label!", "Try again!"

    #--------------------------------------------------------
    
    if all(check_):
        if Jumper(label) != False:
            return f"Label: {Jumper(label)[1]}", f"Result: {Jumper(label)[0]}"  
        elif Three_Digit(label) != False:
            return f"Label: {Three_Digit(label)[1]}", f"Result: {Three_Digit(label)[0]}"  
        elif Four_Digit(label) != False:
            return f"Label: {Four_Digit(label)[1]}", f"Result: {Four_Digit(label)[0]}"    
    else:
        return "Oops, Seems like we have a problem!", "Make sure you have entered the label correctly!!"
    
```

```python
def main():
    """
    
    """

    #--------------------------------------------------------

    user_choice = input("Select the type of your resistor:\n1. Dip 2. SMD\n>> ").strip().lower()
    
    #--------------------------------------------------------

    if user_choice == "dip" or user_choice == "1":
        color, result = Dip_Res()
        return color, result
    elif user_choice == "smd" or user_choice == "2":
        label, result = SMD_Res()
        return label, result
    else:
        return "Your choice is not among the options!", "Please try again!"


c, r = main()
print(c, r, sep="\n")
```
